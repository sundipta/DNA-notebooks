```python
row_numbers = list(df1[df1["Time"] == 0].iloc[1:].index)
offsets = [180]
def fix_Time(row_numbers, offsets, dataframe):
    if len(row_numbers) != len(offsets):
        raise Exception("Fix your vectors")
    
    copy = dataframe.copy()
    
    for row_number, offset in zip(row_numbers, offsets):
        last = int(row_number) - 1
        to_add = dataframe.loc[last]['Time'] + offset
        print to_add
        copy.loc[dataframe[dataframe.index >= row_number].index, 'Time'] += to_add
    
    return copy

df2 = fix_Time(row_numbers, offsets, df1)
df2[df2["Time"]== 0]
```

```python
res = pd.DataFrame({'Time': df2['Time'], 'fluor': df2['A1']})
res['exp']= dict_pr["A1"]["experiment"]
res['cond']= dict_pr["A1"]["treatment"]
res['rep']= dict_pr["A1"]["replicate"]
print res.head()
```

```python
df_list = []
for key, value in sorted(dict_pr.iteritems(), key=lambda x: x[0]):
    res = pd.DataFrame({'Time': df2['Time'], 'fluor': df2[key]})
    res['exp'] = value["experiment"]
    res['cond'] = value["treatment"]
    res['rep'] = value["replicate"]
    res['orig'] = key
    df_list.append(res)
```

```python
df3 = pd.concat(df_list)
df3
```

```python
df_T1 = df3.loc[df3["exp"]== "T1"]
df_T1.reset_index(drop=True, inplace=True)
```

```python
grouped = df_T1.groupby(["cond", "Time"])

df_mean = grouped.mean()
mean = df_mean.reset_index()
std1 = grouped.std().reset_index()
print std1.head()
piv = std1.pivot(index="Time",columns = "cond",values = "fluor")
piv1=mean.pivot(index="Time",columns = "cond",values = "fluor")
#piv1.to_csv('170127_T1_mean.tsv', sep = '\t')
piv1.plot(yerr=piv)
```
