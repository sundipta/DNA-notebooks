```python
new_data = []
for idx, time_str in df1["Time"].iteritems():
    x = time.strptime(time_str,'%H:%M:%S')
    secs = datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds() 
    new_data.append(secs)
df1["Time_secs"] = new_data
```

```python
row_numbers = list(df1[df1["Time_secs"] == 0].iloc[1:].index)
offsets = [0,0]
def fix_Time(row_numbers, offsets, dataframe):
    if len(row_numbers) != len(offsets):
        raise Exception("Fix your vectors")
    
    copy = dataframe.copy()
    
    for row_number, offset in zip(row_numbers, offsets):
        last = int(row_number) - 1
        to_add = dataframe.loc[last]['Time_secs'] + offset
        print to_add
        copy.loc[dataframe[dataframe.index >= row_number].index, 'Time_secs'] += to_add
    
    return copy

df2 = fix_Time(row_numbers, offsets, df1)
df2[df2["Time_secs"]== 0]
```
